## Wprowadzenie do macierzy

Teraz, gdy omówiliśmy wektory, możemy rozszerzyć ten pomysł na koncepcję __macierzy__. Macierze to prostokątne uporządkowane tablice liczb, symboli lub wyrażeń ułożonych w rzędy i kolumny. Są uważane za uporządkowane, ponieważ każda pozycja wejściowa ma inne znaczenie. Często przydaje się traktowanie macierzy jako sekwencji __wektorów kolumn__ lub __wektorów wierszy__ przy określonych operacjach. W rzeczywistości, biorąc pod uwagę naszą definicję, poszczególne wektory można również traktować jako macierze! Rozważmy na przykład trzy macierze poniżej:

$$ A = \begin{bmatrix} 1 \\ 2  \end{bmatrix} \text{, }B = \begin{bmatrix} 1 & 2 & 3\\ 3 & 2 & 1 \end{bmatrix} \text{, } C =\begin{bmatrix} 1 & 2 \\ 2 & 3 \\ 3 & 4 \end{bmatrix}$$

Wymiary macierzy oznaczamy najpierw liczbą wierszy, a następnie liczbą kolumn. W tym przypadku A to macierz 2x1, B to macierz 2x3, a C to macierz 3x2. Macierz A jest również wektorem kolumnowym, a macierz B może być również konceptualizowana jako trzy wektory kolumnowe 2D lub nawet dwa wektory wierszowe 3D! Niech M,N będą dwiema liczbami całkowitymi, gdzie M,N > 1, a $a_{ij}, i=1,...,M, j=1,...,N$ są liczbami $MN$, możemy mieć ogólną reprezentację macierzy w następujący sposób:


$$ A = \begin{bmatrix} a_{11} & ... & a_{1j} & ... & a_{1N} \\
                        \vdots &    & \vdots &     & \vdots \\
                        a_{i1} & ... & a_{ij} & ... & a_{iN} \\
                        \vdots &    & \vdots &     & \vdots \\
                        a_{M1} & ... & a_{Mj} & ... & a_{MN}
\end{bmatrix}$$


Macierze są znaczącym sposobem przedstawiania danych. Ogólnie rzecz biorąc, poruszanie się po wierszach ma określone znaczenie, które różni się od poruszania się po kolumnach. Załóżmy na przykład, że zbieramy dane o populacji i zbieramy wzrost, wagę i płeć od każdej osoby w próbie 8 osób. Prosta reprezentacja macierzowa tych danych polega na tym, że poruszając się po wierszach, przechodzimy do innej funkcji, a poruszając się po kolumnach, przechodzimy do danych innej osoby, jak pokazano na poniższym diagramie. Widzimy, że w tym przypadku po raz kolejny możemy potraktować macierz jako listę wektorów, gdzie każdy wektor zawiera wagę, wzrost i płeć jednej osoby.

## Mnożenie macierzy
Widzieliśmy mnożenie wektorów, możemy też przeprowadzić __mnożenie macierzy__. W tej operacji kolejność jest ważna, a druga macierz musi mieć taką samą liczbę wierszy, ile pierwsza macierz ma kolumn. Biorąc pod uwagę macierz MxN A i macierz NxK B, ich __iloczyn macierzy__, $C = A\times B$, będzie macierzą MxK. Oznacza to, że będzie mieć tyle wierszy, co pierwsza macierz i tyle samo kolumn, co druga macierz. Należy pamiętać, że __do mnożenia macierzy pierwsza macierz musi mieć tyle kolumn, co druga wierszy__. Biorąc pod uwagę ogólną definicję A i B poniżej, ich iloczyn macierzowy jest następujący:

$$ A = \begin{bmatrix} a_{11} & ... & a_{1j} & ... & a_{1N} \\
                        \vdots &    & \vdots &     & \vdots \\
                        a_{i1} & ... & a_{ij} & ... & a_{iN}\\
                        \vdots &    & \vdots &     & \vdots \\
                        a_{M1} & ... & a_{Mj} & ... & a_{MN}
\end{bmatrix}, \;
B = \begin{bmatrix} b_{11} & ... & b_{1j} & ... & b_{1K} \\
                        \vdots &    & \vdots &     & \vdots \\
                        b_{i1} & ... & b_{ij} & ... & b_{iK} \\
                        \vdots &    & \vdots &     & \vdots \\
                        b_{N1} & ... & b_{Nj} & ... & b_{MK}
\end{bmatrix}$$

$$ C = A\times B = \begin{bmatrix} c_{11} & ... & c_{1j} & ... & c_{1K} \\
                        \vdots &    & \vdots &     & \vdots \\
                        c_{i1} & ... & c_{ij} & ... & c_{iK}\\
                        \vdots &    & \vdots &     & \vdots \\
                        c_{M1} & ... & c_{Mj} & ... & c_{MK}
\end{bmatrix}$$

Gdzie:

$$C_{ij} = \sum_{k=1}^{N}a_{ik}b_{kj}$$

Ta ogólna definicja może być trudna do zrozumienia, więc aby uprościć ten proces, będziemy traktować macierze jako listy wektorów kolumnowych i wierszowych. W tym przypadku wpis iloczynu macierzy $c_{ij}$ jest po prostu __iloczynem skalarnym __ $\mathbf{i^{th}}$ __wiersza pierwszej macierzy, A i__ $\mathbf{j ^{th}}$ __kolumna drugiej macierzy, B__. Poniżej przedstawiono dwa przykłady iloczynu macierzy:

$$ \begin{bmatrix} 1 & 2 \\ 3 & 4  \end{bmatrix} \times \begin{bmatrix} 5\\ 6 \end{bmatrix} =
\begin{bmatrix}
\begin{bmatrix}1 & 2 \end{bmatrix}\cdot \begin{bmatrix}5 \\ 6 \end{bmatrix}\\
\begin{bmatrix}3 & 4 \end{bmatrix}\cdot \begin{bmatrix}5 \\ 6 \end{bmatrix}
\end{bmatrix}
= \begin{bmatrix} 5 \\ 4 \end{bmatrix}$$

$$\begin{bmatrix} 2 & 3 \\ 1 & 4 \\ 1 & 0 \end{bmatrix} \times \begin{bmatrix} 1 & 0 \\ 2 & 1 \end{bmatrix} =
\begin{bmatrix}
\begin{bmatrix} 2 & 3 \end{bmatrix} \cdot \begin{bmatrix} 1 \\ 2 \end{bmatrix} &
\begin{bmatrix} 2 & 3 \end{bmatrix} \cdot \begin{bmatrix} 0 \\ 1 \end{bmatrix} \\
\begin{bmatrix} 1 & 4 \end{bmatrix} \cdot \begin{bmatrix} 1 \\ 2 \end{bmatrix} &
\begin{bmatrix} 1 & 4 \end{bmatrix} \cdot \begin{bmatrix} 0 \\ 1 \end{bmatrix} \\
\begin{bmatrix} 1 & 0 \end{bmatrix} \cdot \begin{bmatrix} 1 \\ 2 \end{bmatrix} &
\begin{bmatrix} 1 & 0 \end{bmatrix} \cdot \begin{bmatrix} 0 \\ 1 \end{bmatrix}
\end{bmatrix} =
\begin{bmatrix} 8 & 3 \\ 9 & 4 \\ 1 & 0 \end{bmatrix}
$$

Jak widzieliście powyżej, nawet przy pomocy iloczynu skalarnego, jest to czasochłonne zadanie do wykonania ręcznie i praktycznie niemożliwe w przypadku bardzo dużych macierzy. ## Macierze tożsamościowe
Teraz, gdy omówiliśmy mnożenie macierzy, możemy omówić kilka interesujących właściwości i typów macierzy. Kluczowy typ macierzy to __macież tożsamościowa__. Macierze tożsamościowe muszą być __macierzami kwadratowymi__, co oznacza, że mają tyle wierszy, ile kolumn. Z definicji iloczyn macierzowy dowolnej macierzy kwadratowej i macierzy jednostkowej o tej samej liczbie wierszy i kolumn jest sobą. Ta właściwość sprawia, że jest to użyteczna macierz, gdy mamy do czynienia z macierzą odwrotną, a także wartościami własnymi i wektorami własnymi, które omówimy wkrótce. Macierze tożsamościowe, oznaczone jako __I__, mają jedynki na głównej przekątnej i zera w innych miejscach, jak pokazano poniżej:

$$ \begin{bmatrix}1 & 0 \\ 0 & 1\end{bmatrix}, \; \begin{bmatrix}1 & 0 & 0 \\ 0 & 1 & 0 \\0 & 0 & 1\end{bmatrix}, \;
\begin{bmatrix} 1 & 0 & ... & 0 \\
                0 & 1 & ... & 0 \\
                \vdots &    & \ddots &  \\
                0 & 0 & ... & 1
\end{bmatrix}
$$

$$A\times I = I\times A = A$$

Możesz sam sprawdzić, a zobaczysz, że każda macierz kwadratowa, którą pomnożysz przez macierz jednostkową, da ci oryginalną macierz! Tutaj pokazujemy, jak wygenerować macierz tożsamości w NumPy, wraz z dowodem, że otrzymujemy oryginalną macierz podczas mnożenia z macierzą tożsamościową.

## Macierze przekątne
Innym użytecznym typem macierzy jest __macierz przekątna__, w której wpisy wzdłuż głównej przekątnej są jedynymi wartościami niezerowymi.

$$D = diag(a_{1}, a_{2}, ..., a_{M}) = \begin{bmatrix} a_{1} & 0 & ... & 0 \\
                      0 & a_{2} & ... & 0  \\
                      \vdots & \vdots & \ddots & \vdots \\
                       0 & 0 & ... & a_{M}
\end{bmatrix}$$
Gdzie $a_{i}, i=1,...,M$ są liczbami skalarnymi.

Macierze diagonalne mają ciekawe właściwości:
- kolejność mnożenia macierzy nie ma znaczenia przy macierzach diagonalnych
- transpozycja macierzy diagonalnej jest sobą
- skaluje wartość każdego wiersza macierzy o odpowiednią składową diagonalną w macierzy diagonalnej

Przykłady:
$$\begin{bmatrix}2 & 0 \\ 0 & 3\end{bmatrix}\times \begin{bmatrix}1 & 1 \\ 1 & 1\end{bmatrix} = \begin{bmatrix}2 & 2 \\ 3 & 3\end{bmatrix} $$

$$\begin{bmatrix}2 & 0 \\ 0 & 4\end{bmatrix}\times \begin{bmatrix}1 & 2 & 3\\ 1 & 2 & 3\end{bmatrix} = \begin{bmatrix}2 & 4 & 6 \\ 4 & 8 & 12\end{bmatrix} $$

Po zrozumieniu niektórych jej właściwości możemy nawet potraktować macierz jednostkową jako macierz diagonalną, która skaluje każdy wiersz o 1.

## Transformacje liniowe
Inną użyteczną interpretacją macierzy jest traktowanie ich jako __przekształcenia liniowe__. Oznacza to, że możemy uznać macierze za coś, co po pomnożeniu przez jeden lub wiele wektorów wejściowych otrzymamy jeden lub wiele wektorów wyjściowych w tej samej przestrzeni wektorowej. Rozważmy na przykład poniższy iloczyn mnacierzowy. Widzimy, że mnożąc wektor przez macierz otrzymujemy inny wektor, oba pokazane na poniższym diagramie:


$$ \begin{bmatrix}1 & 2 \\ 3 & 1 \end{bmatrix} \times \begin{bmatrix}1 \\ 1\end{bmatrix} = \begin{bmatrix}3 \\4\end{bmatrix}$$

```python
# Visualization Code

import plotly.graph_objects as go

x1 = [0, 1, 3]
x2 = [0, 1, 4]


fig = go.Figure(data=[go.Scatter(
    x=x1, y=x2,
    mode='markers',
    marker=dict(size=[10, 30, 30],
            color=["black","black", "orange"])
    )])

fig.update_layout(
    title="Linear Transformation",
    xaxis_title="$x_{1}$",
    yaxis_title="$x_{2}$",
)
fig.add_trace(go.Scatter(x=[0, 1], y=[0, 1],marker_color="black", name="Original Vector"))
fig.add_trace(go.Scatter(x=[0, 3], y=[0, 4],marker_color="orange", name="Transformed Vector"))
fig.update_layout(showlegend=True)
fig.show()
```

Transformacje liniowe mogą wpływać na wektor wejściowy poprzez zmianę jego długości i/lub kierunku. Niektóre przekształcenia liniowe _tylko_ zmieniają długość dowolnego wektora, podczas gdy inne _tylko_ zmieniają kierunek dowolnego wektora. Jednak dla każdej transformacji liniowej istnieją pewne specjalne wektory, które są skalowane tylko przez tę transformację liniową.

Zobaczymy później, że to geometryczne zrozumienie macierzy pomoże w radzeniu sobie z wartościami własnymi i wektorami własnymi.

## Wyznacznik i odwrotność macierzy
__Wyznacznik__ to funkcja macierzy kwadratowej, która redukuje ją do jednej liczby. Wyznacznik macierzy A oznaczamy |A| lub det(A). Jeżeli A składa się z jednego elementu a, to |A| = a. Dla macierzy 2x2 jest on dany wzorem:

$$A = \begin{bmatrix}a & b \\ c & d\end{bmatrix}, \; det(A) = \begin{vmatrix}a & b \\ c & d\end{vmatrix} = ad-bc$$

Znalezienie wyznacznika macierzy kwadratowej NxN dla N > 2 może być wykonane przez rekurencyjne usuwanie wierszy i kolumn w celu utworzenia kolejnych mniejszych macierzy, aż wszystkie będą miały wymiary 2 × 2, a następnie zastosowanie powyższej definicji 2x2. Najbardziej podstawową techniką jest __rozszerzenie o wiersz__, które pokazujemy tutaj dla macierzy 3 × 3:


$$A = \begin{bmatrix}a & b & c \\ d & e & f \\ g & h & i\end{bmatrix} \; det(A) = a\begin{vmatrix} e & f \\ h & i\end{vmatrix} - b\begin{vmatrix} d & f \\ g & i\end{vmatrix} + c\begin{vmatrix} d & e \\ g & h\end{vmatrix}$$

W tym przypadku rozwijamy o wiersz 1, co oznacza usunięcie wiersza 1 poprzez kolejne usuwanie kolumny 1, kolumny 2 i kolumny 3, aby utworzyć trzy macierze 2×2. Wyznacznik każdej mniejszej macierzy jest mnożony przez wpis odpowiadający przecięciu usuniętego wiersza i kolumny. Rozszerzenie naprzemiennie dodaje i odejmuje każdy kolejny wyznacznik.


Do tych obliczeń będziemy zazwyczaj używać komputera, zwłaszcza w przypadku większych wymiarów. Więc po co nam wyznacznik? Cóż, jest to przydatne narzędzie do określania, czy macierz kwadratowa ma __odwrotność__, a jeśli tak, jest również używana do obliczania macierzy odwrotnej.

Jeśli mamy macierz kwadratową, A, jej odwrotność $\mathbf{A^{-1}}$ jest zdefiniowana tak, że:
$$ A\times A^{-1} = A^{-1} \times A = I $$

Odwrotnością macierzy kwadratowej jest macierz o tym samym wymiarze, która po pomnożeniu przez pierwotną macierz zwraca macierz jednostkową, jak pokazano powyżej. Dlatego widzimy znaczenie macierzy tożsamości! Jeśli chodzi o transformacje liniowe, jest to transformacja przeciwna: pobiera wektory wyjściowe i zwraca wektory wejściowe. Jednak nie wszystkie macierze kwadratowe mają odwrotność. Macierze, które nie mają odwrotności, są znane jako __macierze osobliwe__ i mają wyznacznik równy 0. Tak więc, aby określić, czy macierz kwadratowa ma odwrotność, po prostu obliczamy jej wyznacznik! Poniżej przedstawiono kilka przykładów tej koncepcji:

$$B = \begin{bmatrix}1 & 2 \\ 2 & 4\end{bmatrix}, \; det(B) = 0,\; \text{(osobliwa, brak odwrotności)} $$

Innym typem macierzy jest __macierz ortogonalna__. Macierze ortogonalne mają interesujący iloczyn, że ich transpozycja jest jednocześnie ich odwrotnością macierzy! Biorąc pod uwagę macierz ortogonalną A:

$$ A \times A^{T} = A^{T} \times A = I \implies A^{T} = A^{-1}$$

To kolejny przykład, dlaczego warto wizualizować macierze jako złożone z wektorów kolumn/wierszy. W tym przypadku macierz jest ortogonalna, jeśli jej kolumny i wiersze są ortogonalnymi wektorami jednostkowymi (ortonormalnymi). Oznacza to, że iloczyn wewnętrzny jej wierszy z kolumnami da 1, jeśli jest to ten sam wektor, a zero w przeciwnym razie. Aby więc sprawdzić, czy macierz jest ortogonalna, wystarczy udowodnić, że iloczyn skalarny różnych kolumn wynosi 0, a ich długości są równe 1. Przykład macierzy ortogonalnej pokazano poniżej.

$$A = \begin{bmatrix}\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \\ -\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}}\end{bmatrix} = \begin{bmatrix} \mathbf{v_{1}} & \mathbf{v_{2}} \end{bmatrix}, \;\; ||\mathbf{v_{1}}|| = 1, ||\mathbf{v_{2}}|| = 1,  \mathbf{v_{1}} \cdot \mathbf{v_{2}} = 0 \implies \text{orthogonal} $$

$$A \times A^{T} =
\begin{bmatrix}\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \\ -\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}}\end{bmatrix}
\times
\begin{bmatrix}\frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}}\end{bmatrix}
 = \begin{bmatrix}1 & 0 \\ 0 & 1 \end{bmatrix} = I$$

Obliczanie odwrotności macierzy jest ogólnie złożonym procesem, co oznacza, że wykorzystanie ortogonalności macierzy pozwala nam uprościć obliczenia ręczne i zmniejszyć złożoność obliczeniową.

## Wartości własne i wektory własne
Jedno z najważniejszych pojęć w algebrze liniowej: __wartości własne__ i __wektory własne__. Biorąc pod uwagę macierz kwadratową, A, jej wektor(y) własny(e), oznaczony(e) przez $\vec{\mathbf{v_{i}}}$ jest/są wektorami(ami) takimi, że:

$$A\vec{\mathbf{v_{i}}} = \lambda_{i}\vec{\mathbf{v_{i}}}$$

Gdzie $\lambda_{i}$ jest odpowiednią wartością własną, skalarem.

__Wektory własne są dowolnymi możliwymi niezerowymi rozwiązaniami__ $\mathbf{\vec{v_{i}}}$ __powyższego równania__. Co to właściwie oznacza? Cóż, jest to jeden z przypadków, w których sensowne jest wizualizowanie macierzy jako transformacji liniowej. Transformację liniową możemy zastosować do dowolnego wektora z odpowiednią liczbą składowych, a dla kilku wektorów transformacja liniowa skaluje je tylko o podaną wartość skalarną (ich wartość własną). To są wektory własne tej macierzy.

Jak więc obliczyć wartości własne i wektory własne macierzy? Niektórzy z was prawdopodobnie już domyślili się, że wektor ze wszystkimi zerowymi składowymi jest zawsze wektorem własnym. Jest to znane jako __trywialne rozwiązanie__, a ponieważ dotyczy wszystkich macierzy niezależnie od odpowiadającej im wartości własnej, nie mówi nam o żadnych interesujących właściwościach poszczególnych macierzy. Dlatego interesuje(ą) nas __nietrywialne rozwiązanie(a)__ dla wektorów własnych macierzy. Wyrażenie ogólne możemy przepisać w poniższym wzorze, co pozwala nam wykonać następujące czynności:

$$A\vec{\mathbf{v_{i}}} = \lambda_{i}I\vec{\mathbf{v_{i}}}$$
$$(A - \lambda_{i}I)\vec{\mathbf{v_{i}}} = 0 $$

Niektórzy z was mogą pomyśleć: „Możemy pomnożyć obie strony równania przez $(A - \lambda_{i}I)^{-1}$, ale to daje trywialne rozwiązanie, które intuicyjnie znaleźliśmy wcześniej. Co zrobimy teraz?”. Cóż, aby $\vec{\mathbf{v_{i}}}$ był wektorem niezerowym, $(A - \lambda_{i}I)$ musi być liczbą pojedynczą, co oznacza, że NIE MOŻE mieć odwrotności. Dlatego jest tak mało „wyjątków”, które prowadzą do kilku wektorów własnych macierzy spośród nieskończenie wielu innych wektorów w tej samej przestrzeni wektorowej. Idąc o krok dalej, możemy użyć warunku determinującego, który napotkaliśmy wcześniej, aby znaleźć wartości własne macierzy:

$$det(A - \lambda_{i}I) = 0$$

Jeśli zachowamy $\lambda_{i}$ jako zmienną, powyższe wyrażenie daje nam równanie wielomianowe, znane jako __wielomian charakterystyczny__ macierzy, a rozwiązanie go da nam wartości własne macierzy! Aby znaleźć odpowiadające im wektory własne, wstawiamy każdą wartość własną do oryginalnego wyrażenia i rozwiązujemy wektor. Prosty przykład 2x2 pokazano poniżej:


$$A = \begin{bmatrix} 1 & 2 \\ 2 & 1 \end{bmatrix} \implies
A - \lambda_{i}I = \begin{bmatrix} 1-\lambda_{i} & 2 \\ 2 & 1-\lambda_{i} \end{bmatrix}$$

$$det(A - \lambda_{i}I) = 1 - 2 \lambda_{i} + \lambda_{i}^{2} - 4 = \lambda_{i}^{2} - 2 \lambda_{i} - 3 = 0$$
$$ \lambda_{1} = -1, \lambda_{2} = 3$$

Teraz możemy znaleźć odpowiednie wektory własne każdej wartości własnej. Zaczniemy od wektora własnego $\lambda_{1}$:
$$\begin{bmatrix} 1 & 2 \\ 2 & 1 \end{bmatrix} \begin{bmatrix}v_{x} \\ v_{y}\end{bmatrix} =
\begin{bmatrix}-v_{x} \\ -v_{y}\end{bmatrix}$$

$$ v_{x} + 2v_{y} = -v_{x}$$
$$ 2v_{x} + v_{y} = -v_{y}$$
Oznacza to, że odpowiadające wektory własne $\lambda_{1}$ są dowolnymi wektorami takimi, że $v_{x} = -v_{y}$. Możemy je uogólnić, aby otrzymać nasze wektory własne (które są wielokrotnościami skalarnymi):

$$\mathbf{\vec{v_{1}}} = t\begin{bmatrix} 1 \\ -1 \end{bmatrix}, \text{where } t \in \Re$$

Ta sama procedura może być zastosowana dla drugiej wartości własnej, co daje następujący wektor własny:
$$\begin{bmatrix} 1 & 2 \\ 2 & 1 \end{bmatrix} \begin{bmatrix}v_{x} \\ v_{y}\end{bmatrix} =
\begin{bmatrix}3v_{x} \\ 3v_{y}\end{bmatrix}$$

$$ v_{x} + 2v_{y} = 3v_{x}$$
$$ 2v_{x} + v_{y} = 3v_{y}$$
$$v_{x}=v{y}$$
$$\mathbf{\vec{v_{1}}} = \alpha\begin{bmatrix} 1 \\ 1 \end{bmatrix}, \text{where } \alpha \in \Re$$
