```python
# basic
import pandas as pd
import csv
import numpy as np
pd.options.mode.chained_assignment = None

# db
import sqlite3 as sql
```

```python
def read_whole_table(con, table_name):
  return pd.read_sql(f'''select * from {table_name}''', con)
```

```python
DB = 'dede-enus.sqlite'
path_to_db = './drive/MyDrive/Nauczanie/202223/PAD/04_pandas2'
```

```python
# nawiązanie połączenia z bazą
con = sql.connect(f'{path_to_db}/{DB}')
cursor = con.cursor()
```

```python
# wyciągnięcie wszystkich widoków
cursor.execute("SELECT name from sqlite_master WHERE type='view';")
cursor.fetchall()
```

```python
# wyciągnięcie wszystkich tabel
cursor.execute("SELECT name from sqlite_master WHERE type='table';")
cursor.fetchall()
```
